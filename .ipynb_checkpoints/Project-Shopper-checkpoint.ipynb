```python
# because we can infer the last level if we know the values of the other levels for categorical vaiable, we will drop one level from each 
# otherwise the data will have problems with multicollinearity 
col_cat_to_drop = ['OperatingSystems_0', 'Browser_0', 'Region_9', 'TrafficType_0','VisitorType_Other','Weekend_True']
shoppers.drop(col_cat_to_drop, axis=1, inplace=True)
```

###  <font color='navy'>Train and Compare ML Models </font>

```python
# set seed for reproducibility
SEED = 1411

# split dataset into training and test set. imbalaced dataset split in stratified fashion
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=SEED, stratify=y)
```

```python
#instantiate the classifiers
clf_adab = AdaBoostClassifier(random_state=SEED) # default base learner is DecisionTree with max_depth=1
clf_rf = RandomForestClassifier(random_state=SEED) # default n_estimators=100
clf_logreg = LogisticRegression(solver='liblinear', random_state=SEED)
clf_svm = SVC()
```

```python
#name of the classifiers in the same order as classifiers
names = ['Ada Boost', 'Random Forest', 'Logistic Regression', 'SVM']

# create a list of classifiers
classifiers = [clf_adab, clf_rf, clf_logreg, clf_svm]
```

```python
# Metrics of interest 
meancv_scores_list = []
accuracy_list = []
precision_list=[]
recall_list =[]
f1score_list=[]
kappa_list =[]
```

```python
# for each classifier in the classifier above do the following
for clf in classifiers:
    
    #perform a 5-fold CV
    cv_scores = cross_val_score(clf,X_train,y_train,cv=5)
    
    #append the mean of the cross val scores to the mean_cv_scores list
    meancv_scores_list.append(cv_scores.mean())
    
    # fit the classifier to the entire training set 
    clf.fit(X_train, y_train)
    
    # predict test values
    y_pred = clf.predict(X_test)
    
    # calculate and append the metrics of interest
    accuracy_list.append(accuracy_score(y_test, y_pred)) 
    precision_list.append(precision_score(y_test, y_pred))
    recall_list.append(recall_score(y_test, y_pred))
    f1score_list.append(f1_score(y_test, y_pred))
    kappa_list.append(cohen_kappa_score(y_test, y_pred))
```

```python
#create a summary dataframe  
df_summary = pd.DataFrame(list(zip(names, meancv_scores_list,accuracy_list,precision_list,recall_list,f1score_list,kappa_list)),
                          columns=['Classifier','Avg CV Score', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'Kappa']
                         )
df_summary
```

*Because the dataset is imbalaced (85% zero and 15% ones), the accuracy is not a great metric to evaluate the classifiers on. We are interested correctly predicting online shoppers who would make a purchase or not make a purchase (true positives and true negatives). Therefore the classifier with the highest F1 Score ( harmonic mean of the precision and recall) is the one that is most useful.  On this measure, Random Forest appears to be the best classifier*

```python
# plot the roc curve
fig, ax = plt.subplots()
plot_roc_curve(clf_adab, X_test, y_test, name='Ada Boost', ax=ax) #specify axis to plot all roc curves on the same axis
plot_roc_curve(clf_rf, X_test, y_test, name='Random Forest', ax=ax)
plot_roc_curve(clf_logreg, X_test, y_test, name='Logistic', ax=ax)
plot_roc_curve(clf_svm, X_test, y_test, name='SVM', ax=ax)
plt.plot([0, 1], [0, 1], 'k--')
plt.title('ROC Curve')
plt.show()
```

```python
# plot precision-recall curves
fig, ax = plt.subplots()
plot_precision_recall_curve(clf_adab, X_test, y_test, name='Ada Boost', ax=ax) #specify axis to plot all roc curves on the same axis
plot_precision_recall_curve(clf_rf, X_test, y_test, name='Random Forest', ax=ax)
plot_precision_recall_curve(clf_logreg, X_test, y_test, name='Logistic', ax=ax)
plot_precision_recall_curve(clf_svm, X_test, y_test, name='SVM', ax=ax)
plt.title('Precision-Recall Curve')
plt.show()
# Average precision (AP) is the the weighted mean of precisions achieved at each threshold
```

*As observed earlier, the area under ROC is almost identical for both Random Forest and Ada Boost. However the precision-recall trade off appears to be much better for the Random Forest classifer*

### Appendix 1 : Dimensionality Reduction with Automatic Recursive Feature Elimination (RFE)

*From the feature importance chart it looks like only a handful of features are contributing most to the prediction. However dropping one weak feature may make other features relatively more or less important.Scikit learn's automatic recursive feature elimination functionality offers a way to minimise the risk of dropping the wrong features by dropping weak features one by one. In this section we will use RFE build a model with only a handful of features and evaluate the model performance*

```python
from sklearn.feature_selection import RFE
```

```python
# Create the RFE with a LogisticRegression estimator and 3 features to select
# 8 => 7 features that seemed to have higher weight (see chart above) plus one additional to account for the combined information captured by the remaining features.
rfe = RFE(estimator=rf_best, n_features_to_select=8, verbose=1)
```

```python
# Fits the eliminator to the data
rfe.fit(X_train, y_train)
```

```python
# Print the features and their ranking (high = dropped early on), 1 =Features Kept
import pprint
pp = pprint.PrettyPrinter()
ranking_dict = dict(zip(X.columns, rfe.ranking_))
ranking_dict = sorted(ranking_dict.items(), key=lambda item: item[1], reverse=True)
pp.pprint(ranking_dict)
```

```python
# Print the features that are not eliminated
print(X.columns[rfe.support_])
```

```python
#test accuracy,f1 score, precision and recall
y_pred_rfe = rfe.predict(X_test)
print('Accuracy :', accuracy_score(y_test, y_pred_rfe))
print('Precision :', precision_score(y_test, y_pred_rfe))
print('Recall :', recall_score(y_test, y_pred_rfe))
print('F1 score :', f1_score(y_test, y_pred_rfe))  
```

```python
# plot confusion matrix normalized
plot_confusion_matrix(rfe, X_test, y_test,cmap=plt.cm.Blues)  
plt.title('Confusion Matrix')
plt.show()
```

```python
# plot confusion matrix normalized
plot_confusion_matrix(rfe, X_test, y_test,cmap=plt.cm.Blues, normalize='true')  
plt.title('Confusion Matrix with Normalization')
plt.show()
```
